```python
# нормируем признаки

import sklearn.preprocessing as sp

scaler = sp.StandardScaler()

x = np.array(df.iloc[:, :-1])
y = np.array(df.iloc[:, -1:])
x = scaler.fit_transform(x)
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=21)
```

```python
# должно вернуть 1

X_train[5:6][0], y_train[5:6][0][0]
```

```python
import sklearn.linear_model as lm
from sklearn.metrics import roc_auc_score, auc, accuracy_score, mean_squared_error, roc_curve, classification_report, precision_recall_curve
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
import math

step = 1000
upper = math.ceil(0.67 * x.shape[0]) + step
lower = 1000
```

#kNN (Classifier) алгоритм

```python
# kNN (n_neighbors=2)

N = [x for x in range(lower, 16000, step)]

ERRs_2NN_c = []

for n in N:
    kNN = KNeighborsClassifier(n_neighbors=2)
    kNN.fit(X_train[:n], y_train[:n])
    y_pred = np.array(kNN.predict(X_test[:n]))
    ERRs_2NN_c.append(mean_squared_error(y_pred, y_test[:n]))

print('AUC: ', roc_auc_score(y_test[:n], y_pred))
pos_feat = np.round(y_pred)
print('Accuracy: ', accuracy_score(y_test[:n], pos_feat))

plt.plot(N, ERRs_2NN_c)
plt.xlabel("Nubmer of elements in the train set")
plt.ylabel("MSE")

plt.show()
```

#Обучение логистической регрессии с помощью градиентного спуска

```python
import time
from sklearn.base import BaseEstimator

class LogReg(BaseEstimator):
    def __init__(self, lambda_2=1.0, tolerance=1e-3, max_iter=25, alpha=1e-2):
        """
        lambda_2: L2 regularization param
        tolerance: for stopping gradient descent
        max_iter: maximum number of steps in gradient descent
        alpha: learning rate
        LgR: fitted logistic regression # добавил самостоятельно, чтобы передать в predict_proba ту жу LgR, что обучалась в fit
        """
        
        self.lambda_2 = lambda_2
        self.tolerance = tolerance # установил поменьше, поскольку у меня очень долго считается градиент с 600 элементами
        self.max_iter = max_iter # установил поменьше по тем же причинам
        self.alpha = alpha # установил шаг чуть побольше по тем же причинам
        self.w = None
        self.loss_history = None
        self.LgR = None
    
    def fit(self, X, y):
        """
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: self
        """
        
        self.loss_history = []
        LgR = lm.LogisticRegression(random_state = 42)
        LgR.fit(X, y)
        self.LgR = LgR
        tmp = LgR.coef_[0] # w0 совпадает с последним элементом в LgR.coef_[0]
        self.w = tmp
        return self
    
    def predict_proba(self, X):

        """
        X: np.array of shape (l, d)
        ---
        output: np.array of shape (l, 2) where
        first column has probabilities of -1
        second column has probabilities of +1 (default)
        """

        if self.w is None or self.LgR is None:
            raise Exception('Not trained yet')            
        LgR = self.LgR
        return LgR.predict_proba(X)
    
    def calc_gradient(self, X, y):
        """
        X: np.array of shape (l, d) (l can be equal to 1 if stochastic)
        y: np.array of shape (l)
        ---
        output: np.array of shape (d)
        """
        
        w = self.w
        grad = np.zeros((X.shape[1],))
        l = y.shape[0]
        for j in range(X.shape[1]):
            sum_ = np.sum([(1. / l) * (-1) * X[i, j] * y[j] / (1 + math.exp(y[i] * np.sum(w * X[i, :]))) for i in range(X.shape[0])])
            # Не пониаю, куда в формуле производной выше мог деться множитель (1. / l), поэтому добавил его здесь
            sum_ += self.lambda_2 * w[j] # добавил производную члена с коэффициентом регуляризации
            grad[j] = sum_
        
#         можно было бы нормировать, но тогда мы не сможем выйти из цикла по параметру расстояния между весами
#         grad = normalize(grad[:,np.newaxis], axis=0).ravel()
        
        return grad

    def calc_loss(self, X, y):
        """
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: float
        """
        
        w = self.w
        l = len(y)
        var = np.sum([(1. / l) * (math.log(1 + math.exp(-y[i] * np.sum(w * X[i, :])))) for i in range(l)])
        var += np.sum([self.lambda_2 * (w[i] ** 2) / 2 for i in range(len(w))])
        self.loss_history = np.append(self.loss_history, var)
        return var
```

```python
# Добавили единичный признак

np.append(X_train, np.array([1] * X_train.shape[0]).reshape((X_train.shape[0], 1)), axis=1)
np.append(X_test, np.array([1] * X_test.shape[0]).reshape((X_test.shape[0], 1)), axis=1)
```

```python
a = LogReg() # объект класса
a.fit(X_train, y_train) # чтобы получить начальный вектор весов (из тестовой выборки с помощью логистической регрессии)
max_iter = a.max_iter # получаем макс. количество шагов
tolerance = a.tolerance # получаем значение "чувствительности"
alpha = a.alpha # получаем значение размера шага
num = 1 # счётчик шагов
while True:
    np.append(a.loss_history, a.calc_loss(X_test, y_test))
    w_before = a.w # инициализирую значение вектора весов
    b = time.time()
    grad = a.calc_gradient(X_test, y_test)
    print('step №{}:'.format(num))
    print('Понадобилось {} секунд, чтобы посчитать градиент:'.format(time.time() - b))
    print('Градиент:{}'.format(grad))
    w_after = w_before - alpha * grad
    a.w = w_after
    if (np.sqrt(np.sum((w_after - w_before) ** 2))) < tolerance or num == max_iter:
        break
    print('Вектора весов:\nw_before: {}\nw_after: {}'.format(w_before, w_after))
    num += 1
```

```python
import copy
# считаем качество
y_score = a.predict_proba(X_test)[:, 1] # вероятность положительного класса

threshold = 0.5 # вручную установили порог

y_score = copy.deepcopy(y_score) # скопировали полноценным образом массив с вероятностями
y_score[y_score < threshold] = 0 # заменили значения на 0, где вероятность принадлежности к отрицательному классу < 0.5
y_score[y_score >= threshold] = 1 # заменили значения на 1, где вероятность принадлежности к положительному классу >= 0.5
y_score = [int(elem) for elem in y_score] # float to int

fpr, tpr, thresholds = roc_curve(y_test, y_score) # возвращает FPR, TPR и использованные пороговые значения при построении кривой
roc_auc = auc(fpr, tpr)
roc_auc == roc_auc_score(y_test, y_score), roc_auc
```

```python
# функция, автоматически подбирающая alpha

def optimizer(X_train, y_train, X_validate, y_validate, alpha_min, alpha_max, RegressionClass):
    N = 100
    alpha_grid = np.linspace(alpha_min, alpha_max, N)
    err = np.zeros((N, 1))
    for i, parameter in enumerate(alpha_grid):
        RegressionClass_ = RegressionClass(alpha=parameter)
        RegressionClass_.fit(X_train, y_train)
        y_pred = RegressionClass_.predict(X_validate)
        err[i] = mean_squared_error(y_validate, y_pred)
    alpha = alpha_grid[err.argmin()]
    
#     alpha_grid - массив перебираемых значений гиперпараметра
#     alpha - найденное оптимальное значение гиперпараметра 
#     err - массив среднеквадратичных отклонений для каждого из alpha_grid
    return alpha, err, alpha_grid
```
